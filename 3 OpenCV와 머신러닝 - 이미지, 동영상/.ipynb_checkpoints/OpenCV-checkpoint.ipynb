# 이미지 읽기 

```python
# 이미지 다운로드
import urllib.request as req
url = "http://uta.pw/shodou/img/28/214.png"
req.urlretrieve(url, "test.png")
```

```python
# openCV로 읽어드이기
import cv2
img = cv2.imread("test.png")
print(img)
```

```python
# 이미지가 없으면 None을 반환
img2 = cv2.imread("존재하지 않는 파일.png")
print(img2)
```

# 이미지 인라인 출력하기

```python
%matplotlib inline
```

```python
# 다운로드한 이미지 출력하기
import matplotlib.pyplot as plt
import cv2

img = cv2.imread("test.png")
plt.axis("off")  # 격자와 숫자 제거
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
