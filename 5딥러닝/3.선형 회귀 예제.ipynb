# 트레이닝 데이터 도식

![트레이닝 데이터](3-1.png)

```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD, Adam

import numpy as np
```

```python
# 데이터 생성
x_data = np.array([[1, 2, 0], [5, 4, 3], [1, 2, -1], [3, 1, 0], [2, 4, 2], 
                   [4, 1, 2], [-1, 3, 2], [4, 3, 3], [0, 2, 6], [2, 2, 1], 
                   [1, -2, -2], [0, 1, 3], [1, 1, 3], [0, 1, 4], [2, 3, 3]])
y_data = np.array([-4, 4, -6, 3, -4,
                    9, -7, 5, 6, 0,
                    4, 3, 5, 5, 1])
#print(x_data.shape, y_data.shape)
```

```python
# 모델 구축
model = Sequential()
model.add(Dense(1, input_shape = (3,), activation='linear'))
```

```python
# 모델 컴파일
model.compile(optimizer=SGD(learning_rate=1e-2), loss='mse')
model.summary()
```

```python
# 모델 학습
hist = model.fit(x_data, y_data, epochs=1000)
```

```python
# 모델 평가 및 예측
test_data = [ [5,5,0], [2,3,1], [-1,0,-1], [10,5,2], [4,-1,-2] ]
ret_val = [2*data[0] - 3 *data[1] + 2*data[2] for data in test_data]

prediction_val = model.predict(np.array(test_data))
print(prediction_val)
print('=========')
print(ret_val)
```

```python
# 모델의 입력과 출력의 데이터 타입과 형상을 알수 있음
print(model.input)
print(model.output)

# 우리가 학습한 시스템의 가중치와 바이어스가 어떤값으로 최적화 되어있는지 확인이 가능하다.
print(model.weights)
```

```python
# 손실 함수 추세
import matplotlib.pyplot as plt

plt.title('Loss Trend')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid()

# 손실함수 값은 fit() 메소드의 리턴 밸류의 히스토리에 자동 저장됨
plt.plot(hist.history['loss'], label = 'train loss')
plt.legend(loc='best')

plt.show()

# 손실 함수는 지속적으로 감소한다. 이상태면 약 150 ~ 180 번 반복하면 많이 정확해진다.
```
