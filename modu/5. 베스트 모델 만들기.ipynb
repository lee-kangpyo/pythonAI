# 포르투갈 서북쪽의 비뉴 베르드 지방의 와인 측정 데이터
## 와인 샘플 1599개, 화이트와인 4898개를 등급 맛, 산도를 측정해 분석
### UCL 저장소에 올라온 각각의 분리된 데이터를 합쳐서 레드와인과 화이트 와인을 구분하는 모델 만들기

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

import pandas as pd
import numpy as np
import tensorflow as tf

import matplotlib.pyplot as plt
```

```python
# seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 데이터 입력
df = pd.read_csv("./dataset/wine.csv", header=None)
df_shuffle = df.sample(frac=1)  # 데이터 샘플링 (frac=1 => 100%, frac=0.5 => 50%)
dataset = df_shuffle.values
x = dataset[:, 0:12]
y = dataset[:, -1]
```

```python
df_shuffle.info()
# 한 row에 13 개의 정보가 있다.
# 6492 개의 샘플이 존재
'''
0 : 주석산 농도          1 : 아세트산 농도
2 : 구연산 농도          3 : 잔류 당분 농도
4 : 염화나트륨 농도      5 : 유리 아황산 농도
6 : 총 아황산 농도       7 : 밀도
8 : PH                   9 : 황산 칼륨 농도
10: 알코올 도수          11: 와인의 맛 (0 ~ 10등급)
12: class (1: 레드와인, 0: 화이트 와인)
'''
```

```python
# 모델 설정 이항분류 이므로 sigmoid 함수 사용
model = Sequential()
model.add(Dense(30, input_dim=12, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

model.summary()
# 12개의 입력값에 4개의 은닉층을 만들어 각각 30, 12, 8, 1 개의 노드를 설정
```

```python
# 모델 컴파일
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# 모델 학습
model.fit(x, y, epochs=200, batch_size=200)
```

```python
#결과 확인
print(f"{model.evaluate(x, y)[1]}")
```

# 모델 업데이트하기 
## 모델을 저장하고 재사용 하는방법은 save()함수와 load_model() 함수를 이용
## 이번에는 각 epoch 마다 모델의 정확도를 함께 기록하면서 저장해보자

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

import pandas as pd
import numpy as np
import tensorflow as tf

import matplotlib.pyplot as plt

import os
```

```python
# seed 값설정
np.random.seed(3)
tf.random.set_seed(3)
```

```python
# 데이터 입력
df_pre = pd.read_csv("./dataset/wine.csv", header=None)
df = df_pre.sample(frac=1)  # 데이터 샘플링 (frac=1 => 100%, frac=0.5 => 50%)

dataset = df.values
x = dataset[:, 0:12]
y = dataset[:, -1]
```

```python
# 모델 설정
model = Sequential()
model.add(Dense(30, input_dim=12, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

```python
# 모델 컴파일
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# 모델 저장 폴더 설정
MODEL_DIR = "./model/"
if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)
```

```python
# 모델 저장 조건 설정
modelpath = "./model/{epoch:02d}-{val_loss:.4f}.hdf5"
checkpointer = ModelCheckpoint(filepath=modelpath, monitor="val_loss", verbose=1, save_best_only=True)
# verbose = 1 [해당함수의 진행사항이 출력], 0 [출력안됨]
# 테스트 오차 : val_loss
# 학습 정확도 : acc
# 테스트셋 정확도 : val_acc
# 학습셋 오차 : loss
```

```python
# 모델 실행 및 저장
model.fit(x,y,validation_split=0.2, epochs=200, batch_size=200, verbose=0, callbacks=[checkpointer])
```

# 그래프로 확인하기
## 학습의 반복이 너무 많으면 과적합이 발생
## 학습의 반복이 너무 적으면 정확한 학습이 안됨
## 모델의 학습 시간에 따른 정확도와 테스트 결과를 그래프로 확인

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

import pandas as pd
import numpy as np
import tensorflow as tf

import matplotlib.pyplot as plt

import os


# seed 값설정
np.random.seed(3)
tf.random.set_seed(3)


# 데이터 입력
df_pre = pd.read_csv("./dataset/wine.csv", header=None)
df = df_pre.sample(frac=1)  # 데이터 샘플링 (frac=1 => 100%, frac=0.5 => 50%)

dataset = df.values
x = dataset[:, 0:12]
y = dataset[:, -1]


# 모델 설정
model = Sequential()
model.add(Dense(30, input_dim=12, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="sigmoid"))


# 모델 컴파일
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])


# 모델 저장 폴더 설정
MODEL_DIR = "./model/"
if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)
    
    
# 모델 저장 조건 설정
modelpath = "./model/{epoch:02d}-{val_loss:.4f}.hdf5"
checkpointer = ModelCheckpoint(filepath=modelpath, monitor="val_loss", verbose=1, save_best_only=True)
```

```python
# 모델 실행 및 저장
hist = model.fit(x, y, validation_split=0.33, epochs = 3500, batch_size=500)
```

```python
#y_vloss에 테스트셋으로 실험 결과의 오차값을 저장
y_vloss = hist.history['val_loss']
```

```python
# y_acc에 학습셋으로 측정한 정확도의 값을 저장
y_acc = hist.history['accuracy']
```

```python
# x값을 지정하고 정확도를 파란색으로 오차를 빨간색으로 표시
x_len = np.arange(len(y_acc))
plt.plot(x_len, y_vloss, "o", c="red", markersize=3)
plt.plot(x_len, y_acc, "o", c="blue", markersize=3)

# 학습셋의 정확도는 시간이 흐를수록 좋아진다 하지만 테스트 결과는 어느정도 이상 시간이 흐르면 더나아지지 않는것을 확인이됨
```
