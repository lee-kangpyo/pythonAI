```python
import pandas as pd
# 데이터 읽어 오기
df = pd.read_csv('./dataset/pima-indians-diabetes.csv', 
                 names=["pregnant", "plasma", "pressure", "thickness", "insulin", "BMI", "pedigree", "age", "class"])

df.head(5)
```

```python
# 데이터 프레임 정보 확인
df.info()
df.describe()
```

```python
# 데이터의 일부 컬럼만 확인
df[["pregnant", "class"]]
```

```python
# 데이터 가공하기 - 임신 횟수당 당뇨병 발생확률
print(df[['pregnant', 'class']].groupby(['pregnant'], as_index=False).mean().sort_values(by='pregnant', ascending=True))
```

```python
# 그래프로 시각화
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,12))
# 히트맵으로 각 항목간의 상관관계를 확인 상관없으면 0에 가까워지고 비슷한 패턴을 변하면 1에 가깝게 변화
sns.heatmap(df.corr(), linewidths=0.1, vmax=0.5, cmap=plt.cm.gist_heat, linecolor='white', annot=True)
plt.show()
```

## df.corr() - 각 항목간의 상관관계를 나타내는 표 생성
## linewidths - 각 그래프의 마진
## vmax - 색상 밝기 조절 높을수로 어두워짐 적절하게 조절해서 가독성을 좋게 해줌
## cmap - 미리 설정된 색상 값을 불러옴
## class(당뇨 발생 여부) 항목을 보면 plasma(공복 혈당 농도) 가 가장 높은 관계를 가진다

```python
# plasma와 class 항목만 떼어서 관계를 그래프로 그리면 아래와 같다.
grid = sns.FacetGrid(df, col='class')
grid.map(plt.hist, 'plasma', bins=10)
plt.show()
# 당뇨 환자의 경우 공복 혈당 농도가 150 이상인 경우가 많음을 알수 있다.
```

# 피마 인디언의 당뇨병 예측 실행

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# seed 값 생성
np.random.seed(3)
tf.random.set_seed(3)
```

```python
# 데이터 로드
dataset = numpy.loadtxt("./dataset/pima-indians-diabetes.csv", delimiter=",")
x = dataset[:, :8]
y = dataset[:, -1]
```

```python
# 모델 설정
model = Sequential()
model.add(Dense(12, input_dim = 8, activation = "relu"))
model.add(Dense(8, activation = "relu"))
model.add(Dense(1, activation = "sigmoid"))
model.summary()
```

```python
# 모델 컴파일
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=["accuracy"])
```

```python
# 모델 실행
model.fit(x, y, epochs=200, batch_size=10)
```

```python
print(f"Accuracy : {model.evaluate(x, y)[1]}")
# 77 퍼센트 정확도를 보인다.
```
