# 소프트 맥스
## 입력값의 총합이 1이 되는 형태로 변경
### 입력 --(가중합)-->[3, 1.5, 0.3] --(소프트맥스)-->[0.7, 0.2, 0.1] --(교차엔트로피)-->[1, 0, 0]
## 큰값은 두드러지게 나타나고 작은값은 더작아지는 효과

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
# seed 값 설정
np.random.seed(3)
tf.random.set_seed(3)
```

```python
# 데이터 입력
df = pd.read_csv("./dataset/iris.csv", names = ["꽃밫침 길이", "꽃받침 너비", "꽃잎 길이", "꽃잎 너비", "species"])
```

```python
# 그래프로 확인
sns.pairplot(df, hue="species")
plt.show()
```

```python
# 데이터 분류
dataset = df.values
x = dataset[:,:4].astype(float)
y_origin = dataset[:, -1]
```

```python
# 문자열을 숫자로 변환
e = LabelEncoder()
e.fit(y_origin)
y = e.transform(y_origin)
y_encoded = tf.keras.utils.to_categorical(y)
```

```python
# 모델설정
model = Sequential()
model.add(Dense(16, input_dim = 4, activation="relu"))   # 입력값 4개를 받고 16개의 노드가 있는 층을 만든다.
model.add(Dense(3, activation = "softmax"))              # 3개의 결과를 반환하는 출력층을 만든다.
model.summary()

# [4개 값] -> 16개 노드 -> [1,0,0] 소프트맥스와 크로스 엔트로피로 원핫 인코딩 값
```

```python
# 모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer="adam", metrics=["accuracy"])
```

```python
# 모델 실행
model.fit(x, y_encoded, epochs=50, batch_size=1)
```

```python
# 결과 출력
print(f"Accuracy : {model.evaluate(x, y_encoded)[1]}")
# 정확도 98%
```
